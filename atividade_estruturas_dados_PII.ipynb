### Exercício 7: Ordenação Multinível
Em Ciência de Dados, frequentemente ordenamos por duas colunas (Ex: Data e Valor). 
Crie uma lista de tuplas `vendas = [("2023-01", 100), ("2023-01", 50), ("2022-12", 200)]`.
Use o **Timsort Nativo** do Python para ordenar primeiro por data e depois por valor. Explique o que significa um algoritmo de ordenação ser "estável".

```python
# Breno

vendas = [("2023-01", 100), ("2023-01", 50), ("2022-12", 200)]
print("Original:", vendas)

por_data_e_valor = sorted(vendas)
print("Por data:", por_data_e_valor)

por_valor = sorted(vendas, key=lambda v: v[1])
print("Por valor:", por_valor)

```

RESPOSTA: Um algoritmo de ordenação é estável quando elementos com chaves iguais mantêm a mesma ordem relativa que tinham na lista original.
